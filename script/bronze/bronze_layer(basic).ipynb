#Read From CSV and Write Silver Table

##Define the Broze Layer DB 

```python
%sql
-- SHOW CATALOGS 
-- show schemas in `abhi-de-ete-project-catlog`
CREATE SCHEMA IF NOT EXISTS `abhi-de-ete-project-catlog`.`bronze`
```

##Define the external location path

```python
#extn_loc has the raw data which was dumped and has to ingest into the bronze layer as is.

extn_loc ='abfss://de-ete-project@abhistgacct.dfs.core.windows.net/'
dbutils.fs.ls(extn_loc)
```

## load "cust_info.csv"

```python
df = (spark.read
      .option
      ('header', True)
      .option
      ('inferSchema', True)
      .csv
      (extn_loc + 'inbound/source_crm/cust_info.csv')
      )
```

```python
df.show(5)
```

```python
df.write.mode("overwrite").saveAsTable("`abhi-de-ete-project-catlog`.bronze.crm_cust_info")
```

## load "prd_info.csv"

```python
df = spark.read.option("header", "true").option("inferSchema", "true").csv(extn_loc + 'inbound/source_crm/prd_info.csv')
df.write.mode("overwrite").saveAsTable("`abhi-de-ete-project-catlog`.bronze.crm_prd_info")
```

## load "sales_details.csv"

```python
df = spark.read.csv(extn_loc+'inbound/source_crm/sales_details.csv', header=True,inferSchema=True)
df.write.mode("overwrite").saveAsTable("`abhi-de-ete-project-catlog`.bronze.crm_sales_details")
```
